````python
import cv2 as cv
import urllib
import numpy as np
from google.colab.patches import cv2_imshow as cv_imshow

# Read an image
def read_image_by_url(url):
    req = urllib.request.urlopen(url)
    arr = np.asarray(bytearray(req.read()), dtype=np.uint8)
    img = cv.imdecode(arr, -1)
    return img

url = "https://th.bing.com/th/id/R.59e124ee288b8d364210765d460f87fb?rik=Fz%2fEGPYIxzy5aA&riu=http%3a%2f%2finvestments.academic.ru%2fpictures%2finvestments%2fimg1965720_CHtenie_knigi.jpg&ehk=%2bFxeKoJorgXma5mfvHxVwCfdijcAlmO2lDYwl6HmSUA%3d&risl=&pid=ImgRaw&r=0"

img = read_image_by_url(url)

cv_imshow(img)


````

````python
rows, cols, _ = img.shape
img = cv.resize(img, (cols//4, rows//4))
cv_imshow(img)
````

````python
cols = cols//4
rows = rows//4
````

````python

M1 = np.float32([[1/2, 0, 0],
[0, 3, 0],
[0, 0, 1]])

dst = cv.warpPerspective(img, M1, (cols, rows*3))
cv_imshow(dst)
````

````python
M2 = np.float32([[-1, 0, rows],
[0, -1, cols],
[0, 0, 1]])

dst = cv.warpPerspective(img, M2, (cols, rows))
cv_imshow(dst)
````

````python
M3 = np.float32([[1, 0, -3],
[0, 1, 1],
[0, 0, 1]])

dst = cv.warpPerspective(img, M3, (cols, rows))
cv_imshow(dst)
````

````python
M4 = np.float32([[1, 0, 0],
[np.sqrt(3), 1, 0],
[0, 0, 1]])

dst = cv.warpPerspective(img, M4, (cols, rows*3))
cv_imshow(dst)
````

````python
M5 = np.float32([[np.sqrt(3)/2, -1/2, 0],
[1/2, np.sqrt(3)/2, 0],
[0, 0, 1]])

dst = cv.warpPerspective(img, M5, (cols, rows))
cv_imshow(dst)
````

````python
M6 = M5@(M4@(M2@M1))
dst = cv.warpPerspective(img, M6, (cols, rows))
cv_imshow(dst)
````
