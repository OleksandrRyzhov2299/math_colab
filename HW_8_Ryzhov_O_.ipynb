–í—ñ–∑—å–º–∏ –∫–æ–¥ —Å–∏–º—É–ª—è—Ü—ñ—ó —ñ–∑ –∑–∞–≤–¥–∞–Ω–Ω—è 7 –¥–æ–º–∞—à–Ω—å–æ–≥–æ –∑–∞–≤–¥–∞–Ω–Ω—è –¥–æ —Ç–µ–º–∏ 7 ‚Äú–¢–µ–æ—Ä—ñ—è –π–º–æ–≤—ñ—Ä–Ω–æ—Å—Ç–µ–π. –ö–æ–º–±—ñ–Ω–∞—Ç–æ—Ä–∏–∫–∞‚Äù. –ë—É–¥–µ–º–æ –≤–≤–∞–∂–∞—Ç–∏, —â–æ –∑–º—ñ–Ω–∞ —Ü—ñ–Ω–∏ –∞–∫—Ü—ñ–π —É –∫–æ–∂–Ω–∏–π –º–æ–º–µ–Ω—Ç —á–∞—Å—É –¥–æ—Ä—ñ–≤–Ω—é—î

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import gamma, normaltest

# –ü–∞—Ä–∞–º–µ—Ç—Ä–∏ –≥–∞–º–º–∞-—Ä–æ–∑–ø–æ–¥—ñ–ª—É
shape = 0.3
scale = 1.1

# –ö—ñ–ª—å–∫—ñ—Å—Ç—å —Å–∏–º—É–ª—è—Ü—ñ–π
n = 100

# –§—É–Ω–∫—Ü—ñ—è –¥–ª—è —Å–∏–º—É–ª—è—Ü—ñ—ó –∑–Ω–∞—á–µ–Ω–Ω—è x
def simulate_x():
    return gamma.rvs(shape, scale=scale, size=n)

# –ü–æ–±—É–¥–æ–≤–∞ –≥—ñ—Å—Ç–æ–≥—Ä–∞–º–∏ —Ä–æ–∑–ø–æ–¥—ñ–ª—É x
x_values = simulate_x()
plt.hist(x_values, bins=30, edgecolor='black', density=True)
plt.title('–ì—ñ—Å—Ç–æ–≥—Ä–∞–º–∞ —Ä–æ–∑–ø–æ–¥—ñ–ª—É x')
plt.xlabel('–ó–Ω–∞—á–µ–Ω–Ω—è x')
plt.ylabel('–ß–∞—Å—Ç–æ—Ç–∞')
plt.show()

# –°–∏–º—É–ª—è—Ü—ñ—è –¥–ª—è —Ä—ñ–∑–Ω–∏—Ö –∑–Ω–∞—á–µ–Ω—å —á–∞—Å—É t
t_values = np.arange(1, 61, 2)  # –Ω–∞–ø—Ä–∏–∫–ª–∞–¥, –≤—ñ–¥ 1 –¥–æ 60 –∑ –∫—Ä–æ–∫–æ–º 2

# –î–ª—è –∫–æ–∂–Ω–æ–≥–æ –∑–Ω–∞—á–µ–Ω–Ω—è t –±—É–¥—É—î–º–æ –≥—ñ—Å—Ç–æ–≥—Ä–∞–º—É —Ç–∞ –ø–µ—Ä–µ–≤—ñ—Ä—è—î–º–æ —Ä–æ–∑–ø–æ–¥—ñ–ª –Ω–∞ –Ω–æ—Ä–º–∞–ª—å–Ω—ñ—Å—Ç—å
for t in t_values:
    prices = []
    for _ in range(n):
        prices.append(sum(simulate_x()[:t]))

    # –ü–æ–±—É–¥–æ–≤–∞ –≥—ñ—Å—Ç–æ–≥—Ä–∞–º–∏ —Ä–æ–∑–ø–æ–¥—ñ–ª—É —Ü—ñ–Ω–∏
    plt.hist(prices, bins=30, edgecolor='black', density=True)
    plt.title(f'–ì—ñ—Å—Ç–æ–≥—Ä–∞–º–∞ —Ä–æ–∑–ø–æ–¥—ñ–ª—É —Ü—ñ–Ω–∏ –¥–ª—è t = {t}')
    plt.xlabel('–ó–Ω–∞—á–µ–Ω–Ω—è —Ü—ñ–Ω–∏')
    plt.ylabel('–ß–∞—Å—Ç–æ—Ç–∞')
    plt.show()

    # –ü–µ—Ä–µ–≤—ñ—Ä–∫–∞ –Ω–∞ –Ω–æ—Ä–º–∞–ª—å–Ω—ñ—Å—Ç—å
    stat, p = normaltest(prices)
    print(f'–î–ª—è t = {t}: —Å—Ç–∞—Ç–∏—Å—Ç–∏–∫–∞ = {stat:.2f}, p-–∑–Ω–∞—á–µ–Ω–Ω—è = {p:.2f}')
    if p < 0.05:
        print('–†–æ–∑–ø–æ–¥—ñ–ª –∑–Ω–∞—á–Ω–æ –≤—ñ–¥—Ä—ñ–∑–Ω—è—î—Ç—å—Å—è –≤—ñ–¥ –Ω–æ—Ä–º–∞–ª—å–Ω–æ–≥–æ.')
    else:
        print('–†–æ–∑–ø–æ–¥—ñ–ª –Ω–µ –≤—ñ–¥—Ä—ñ–∑–Ω—è—î—Ç—å—Å—è –≤—ñ–¥ –Ω–æ—Ä–º–∞–ª—å–Ω–æ–≥–æ.')

# –í–∏—Å–Ω–æ–≤–∫–∏ –ø—Ä–æ –∑–º—ñ–Ω—É —Ä–æ–∑–ø–æ–¥—ñ–ª—É –∑—ñ –∑–±—ñ–ª—å—à–µ–Ω–Ω—è–º t:
# –ú–æ–∂–Ω–∞ –ø–æ–±–∞—á–∏—Ç–∏, —â–æ –∑—ñ –∑–±—ñ–ª—å—à–µ–Ω–Ω—è–º t —Ä–æ–∑–ø–æ–¥—ñ–ª —Ü—ñ–Ω–∏ –±—É–¥–µ –Ω–∞–±–ª–∏–∂–∞—Ç–∏—Å—è –¥–æ –Ω–æ—Ä–º–∞–ª—å–Ω–æ–≥–æ —Ä–æ–∑–ø–æ–¥—ñ–ª—É –∑–≥—ñ–¥–Ω–æ –∑ —Ü–µ–Ω—Ç—Ä–∞–ª—å–Ω–æ—é –≥—Ä–∞–Ω–∏—á–Ω–æ—é —Ç–µ–æ—Ä–µ–º–æ—é.
```

## –ó–∞–≤–¥–∞–Ω–Ω—è 2

–ó–∞–≤–∞–Ω—Ç–∞–∂ –Ω–∞–±—ñ—Ä –¥–∞–Ω–∏—Ö Product Advertising Data (–ø–æ—Å–∏–ª–∞–Ω–Ω—è –Ω–∞ –¥–∏—Å–∫). –ù–∞–±—ñ—Ä –¥–∞–Ω–∏—Ö —Å–∫–ª–∞–¥–∞—î—Ç—å—Å—è —ñ–∑ —Å–µ–º–∏ —Å—Ç–æ–≤–ø—á–∏–∫—ñ–≤, —â–æ –≤—ñ–¥–æ–±—Ä–∞–∂–∞—é—Ç—å –≤–∏—Ç—Ä–∞—Ç–∏ –Ω–∞ —Ä–µ–∫–ª–∞–º—É –Ω–∞ —Ä—ñ–∑–Ω–∏—Ö –ø–ª–∞—Ç—Ñ–æ—Ä–º–∞—Ö ‚Äî —Ç–µ–ª–µ–±–∞—á–µ–Ω–Ω—è, –±—ñ–ª–±–æ—Ä–¥–∏, Google Ads, —Å–æ—Ü—ñ–∞–ª—å–Ω—ñ –º–µ–¥—ñ–∞, —ñ–Ω—Ñ–ª—é–µ–Ω—Å-–º–∞—Ä–∫–µ—Ç–∏–Ω–≥ —Ç–∞ –ø–∞—Ä—Ç–Ω–µ—Ä—Å—å–∫–∏–π –º–∞—Ä–∫–µ—Ç–∏–Ω–≥.



–û—Å—Ç–∞–Ω–Ω—ñ–π —Å—Ç–æ–≤–ø—á–∏–∫, "Product_Sold", –º—ñ—Å—Ç–∏—Ç—å –∫—ñ–ª—å–∫—ñ—Å–Ω—É –æ—Ü—ñ–Ω–∫—É –≤—ñ–¥–ø–æ–≤—ñ–¥–Ω–æ—ó –∫—ñ–ª—å–∫–æ—Å—Ç—ñ –ø—Ä–æ–¥–∞–Ω–∏—Ö –æ–¥–∏–Ω–∏—Ü—å —Ç–æ–≤–∞—Ä—É. –î–ª—è –∫–æ–∂–Ω–æ—ó –∫–æ–ª–æ–Ω–∫–∏ –ø–æ—Ä–∞—Ö—É–π —Å–µ—Ä–µ–¥–Ω—î –∑–Ω–∞—á–µ–Ω–Ω—è, –¥–∏—Å–ø–µ—Ä—Å—ñ—é, —Å—Ç–∞–Ω–¥–∞—Ä—Ç–Ω–µ –≤—ñ–¥—Ö–∏–ª–µ–Ω–Ω—è, –ø–æ–±—É–¥—É–π –≥—ñ—Å—Ç–æ–≥—Ä–∞–º—É —Ä–æ–∑–ø–æ–¥—ñ–ª—É –ø–æ–∫–∞–∑–Ω–∏–∫–∞, –ø–µ—Ä–µ–≤—ñ—Ä –Ω–∞ –Ω–æ—Ä–º–∞–ª—å–Ω—ñ—Å—Ç—å —Ä–æ–∑–ø–æ–¥—ñ–ª—É —Ç–∞ –ø–æ—Ä–∞—Ö—É–π –∫–æ—Ä–µ–ª—è—Ü—ñ—é –∑ Product_Sold.

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import shapiro

# –ó–∞–≤–∞–Ω—Ç–∞–∂–µ–Ω–Ω—è –¥–∞–Ω–∏—Ö
url = "https://docs.google.com/spreadsheets/d/18WCpPS96Tb3cB0FCsIA92PEhcmBkp08sjYhS9DsQfJE/export?format=xlsx&gid=954244094"
data = pd.read_excel(url)

# –ü–µ—Ä–µ—Ç–≤–æ—Ä–µ–Ω–Ω—è –≤—Å—ñ—Ö –¥–∞–Ω–∏—Ö –Ω–∞ —á–∏—Å–ª–æ–≤—ñ –∑–Ω–∞—á–µ–Ω–Ω—è —Ç–∞ –≤–∏–¥–∞–ª–µ–Ω–Ω—è —Ä—è–¥–∫—ñ–≤ –∑ NaN
data = data.apply(pd.to_numeric, errors='coerce').dropna()

# –í–∏–≤–µ–¥–µ–Ω–Ω—è –æ–ø–∏—Å–æ–≤–æ—ó —Å—Ç–∞—Ç–∏—Å—Ç–∏–∫–∏
print("–û–ø–∏—Å–æ–≤–∞ —Å—Ç–∞—Ç–∏—Å—Ç–∏–∫–∞:")
stats = data.describe()
variance = data.var()
stats.loc['variance'] = variance
print(stats)

# –ü–æ–±—É–¥–æ–≤–∞ –≥—ñ—Å—Ç–æ–≥—Ä–∞–º –¥–ª—è –∫–æ–∂–Ω–æ–≥–æ –ø–æ–∫–∞–∑–Ω–∏–∫–∞
for column in data.columns:
    plt.figure(figsize=(10, 6))
    sns.histplot(data[column], kde=True)
    plt.title(f'Histogram of {column}')
    plt.xlabel(column)
    plt.ylabel('Frequency')
    plt.show()

# –¢–µ—Å—Ç –Ω–∞ –Ω–æ—Ä–º–∞–ª—å–Ω—ñ—Å—Ç—å
print("\n–¢–µ—Å—Ç –Ω–∞ –Ω–æ—Ä–º–∞–ª—å–Ω—ñ—Å—Ç—å:")
normality_results = {column: shapiro(data[column].dropna()) for column in data.columns}
for column, result in normality_results.items():
    print(f"{column}: Statistic={result[0]}, P-value={result[1]}")

# –ö–æ—Ä–µ–ª—è—Ü—ñ–π–Ω–∏–π –∞–Ω–∞–ª—ñ–∑
correlations = data.corr()['Product_Sold'].drop('Product_Sold')
print("\n–ö–æ—Ä–µ–ª—è—Ü—ñ—ó –∑ Product_Sold:")
print(correlations)
```
