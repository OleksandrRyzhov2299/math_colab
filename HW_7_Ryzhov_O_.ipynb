–£ –ø–∞—Ä—Ç—ñ—ó –∑ 50 –¥–µ—Ç–∞–ª–µ–π 5 –Ω–µ—Å—Ç–∞–Ω–¥–∞—Ä—Ç–Ω–∏—Ö. –ó–Ω–∞–π–¥–µ–º–æ –π–º–æ–≤—ñ—Ä–Ω—ñ—Å—Ç—å —Ç–æ–≥–æ, —â–æ —Å–µ—Ä–µ–¥ –æ–±—Ä–∞–Ω–∏—Ö –Ω–∞–≤–º–∞–Ω–Ω—è –¥–ª—è –ø–µ—Ä–µ–≤—ñ—Ä–∫–∏ —à–µ—Å—Ç–∏ –¥–µ—Ç–∞–ª–µ–π –¥–≤—ñ –≤–∏—è–≤–ª—è—Ç—å—Å—è –Ω–µ—Å—Ç–∞–Ω–¥–∞—Ä—Ç–Ω–∏–º–∏.

#### –†–æ–∑–≤'—è–∑–∞–Ω–Ω—è:

–ô–º–æ–≤—ñ—Ä–Ω—ñ—Å—Ç—å –ø–æ–¥—ñ—ó –æ–±—á–∏—Å–ª—é—î—Ç—å—Å—è —è–∫ –≤—ñ–¥–Ω–æ—à–µ–Ω–Ω—è –∫—ñ–ª—å–∫–æ—Å—Ç—ñ —Å–ø—Ä–∏—è—Ç–ª–∏–≤–∏—Ö –≤–∏–ø–∞–¥–∫—ñ–≤ –¥–æ –∑–∞–≥–∞–ª—å–Ω–æ—ó –∫—ñ–ª—å–∫–æ—Å—Ç—ñ –º–æ–∂–ª–∏–≤–∏—Ö –≤–∏–ø–∞–¥–∫—ñ–≤:

1. **–ö—ñ–ª—å–∫—ñ—Å—Ç—å —Å–ø—Ä–∏—è—Ç–ª–∏–≤–∏—Ö –≤–∏–ø–∞–¥–∫—ñ–≤**:
   –î–ª—è —Ç–æ–≥–æ, —â–æ–± 2 —ñ–∑ 6 –æ–±—Ä–∞–Ω–∏—Ö –¥–µ—Ç–∞–ª–µ–π –±—É–ª–∏ –Ω–µ—Å—Ç–∞–Ω–¥–∞—Ä—Ç–Ω–∏–º–∏, —Å–ø–µ—Ä—à—É –ø–æ—Ç—Ä—ñ–±–Ω–æ –≤–∏–±—Ä–∞—Ç–∏ 2 –Ω–µ—Å—Ç–∞–Ω–¥–∞—Ä—Ç–Ω—ñ –¥–µ—Ç–∞–ª—ñ —ñ–∑ –Ω–∞—è–≤–Ω–∏—Ö 5. –¶–µ –º–æ–∂–Ω–∞ –∑—Ä–æ–±–∏—Ç–∏ –∑–∞ –¥–æ–ø–æ–º–æ–≥–æ—é –∫–æ–º–±—ñ–Ω–∞—Ü—ñ—ó $ \binom{5}{2} $.

2. **–ö—ñ–ª—å–∫—ñ—Å—Ç—å –≤—Å—ñ—Ö –º–æ–∂–ª–∏–≤–∏—Ö –≤–∏–ø–∞–¥–∫—ñ–≤**:
   –í–∏–±—ñ—Ä 6 –¥–µ—Ç–∞–ª–µ–π —ñ–∑ 50 –º–æ–∂–ª–∏–≤–∏–π –∑–∞ –¥–æ–ø–æ–º–æ–≥–æ—é –∫–æ–º–±—ñ–Ω–∞—Ü—ñ—ó $ \binom{50}{6} $.

–û—Ç–∂–µ, –π–º–æ–≤—ñ—Ä–Ω—ñ—Å—Ç—å —Ç–æ–≥–æ, —â–æ —Å–µ—Ä–µ–¥ –æ–±—Ä–∞–Ω–∏—Ö –Ω–∞—É–≥–∞–¥ —à–µ—Å—Ç–∏ –¥–µ—Ç–∞–ª–µ–π –¥–≤—ñ –±—É–¥—É—Ç—å –Ω–µ—Å—Ç–∞–Ω–¥–∞—Ä—Ç–Ω–∏–º–∏, –æ–±—á–∏—Å–ª—é—î—Ç—å—Å—è —è–∫:

$ P = \frac{\binom{5}{2}}{\binom{50}{6}} $

–î–µ $ \binom{5}{2} $ - —Ü–µ —á–∏—Å–ª–æ —Å–ø–æ—Å–æ–±—ñ–≤ –≤–∏–±—Ä–∞—Ç–∏ 2 –Ω–µ—Å—Ç–∞–Ω–¥–∞—Ä—Ç–Ω—ñ –¥–µ—Ç–∞–ª—ñ –∑ 5, –∞ $ \binom{50}{6} $ - —á–∏—Å–ª–æ —Å–ø–æ—Å–æ–±—ñ–≤ –≤–∏–±—Ä–∞—Ç–∏ –±—É–¥—å-—è–∫—ñ 6 –¥–µ—Ç–∞–ª–µ–π –∑ 50.

–û–±—á–∏—Å–ª–∏–≤—à–∏ $ \binom{5}{2} = 10 $ —Ç–∞ $ \binom{50}{6} $, –º–æ–∂–Ω–∞ –∑–Ω–∞–π—Ç–∏ –π–º–æ–≤—ñ—Ä–Ω—ñ—Å—Ç—å $ P $.

–¶–µ–π –ø—ñ–¥—Ö—ñ–¥ –¥–æ–∑–≤–æ–ª—è—î –∞–Ω–∞–ª—ñ—Ç–∏—á–Ω–æ –∑–Ω–∞–π—Ç–∏ —Ç–æ—á–Ω—É –π–º–æ–≤—ñ—Ä–Ω—ñ—Å—Ç—å –±–µ–∑ –ø–æ—Ç—Ä–µ–±–∏ —É –∫–æ–Ω–∫—Ä–µ—Ç–Ω–∏—Ö —á–∏—Å–ª–æ–≤–∏—Ö –∑–Ω–∞—á–µ–Ω–Ω—è—Ö.


## –ó–∞–≤–¥–∞–Ω–Ω—è 2
–î–ª—è –∑–∞–¥–∞—á—ñ –∑ –≤–∏–±—ñ—Ä–∫–æ—é 5 —Ä–µ–π—Å—ñ–≤ —ñ–∑ 15-—Ç–∏, –¥–µ 60% —Ä–µ–π—Å—ñ–≤ –≤–∏–∫–æ–Ω—É—é—Ç—å—Å—è –Ω–∞ –≤–ª–∞—Å–Ω–æ–º—É –ø–∞—Ä–∫—É, —à—É–∫–∞—î–º–æ –π–º–æ–≤—ñ—Ä–Ω—ñ—Å—Ç—å, —â–æ —Å–∞–º–µ 3 –∑ —Ü–∏—Ö 5 —Ä–µ–π—Å—ñ–≤ –≤–∏–∫–æ–Ω—É—é—Ç—å—Å—è –Ω–∞ –≤–ª–∞—Å–Ω–æ–º—É –ø–∞—Ä–∫—É.

–ú–∞—î–º–æ:
- –ö—ñ–ª—å–∫—ñ—Å—Ç—å —Ä–µ–π—Å—ñ–≤: $ n = 15$
- –ô–º–æ–≤—ñ—Ä–Ω—ñ—Å—Ç—å —Ä–µ–π—Å—É –Ω–∞ –≤–ª–∞—Å–Ω–æ–º—É –ø–∞—Ä–∫—É:  $p = 0.6$
- –ö—ñ–ª—å–∫—ñ—Å—Ç—å –≤–∏–±—Ä–∞–Ω–∏—Ö —Ä–µ–π—Å—ñ–≤:  $k = 5 $
- –®—É–∫–∞—î–º–æ –π–º–æ–≤—ñ—Ä–Ω—ñ—Å—Ç—å  $$PX = 3$$ , –¥–µ \ X \ ‚Äî –∫—ñ–ª—å–∫—ñ—Å—Ç—å —Ä–µ–π—Å—ñ–≤ –Ω–∞ –≤–ª–∞—Å–Ω–æ–º—É –ø–∞—Ä–∫—É —Å–µ—Ä–µ–¥ 5-—Ç–∏.

–í–∏–∫–æ—Ä–∏—Å—Ç–æ–≤—É—î–º–æ –±—ñ–Ω–æ–º—ñ–∞–ª—å–Ω–∏–π —Ä–æ–∑–ø–æ–¥—ñ–ª:

 $ PX = 3 = \binom{5}{3} \cdot 0.6^3 \cdot 0.4^2 \ $

–û–±—á–∏—Å–ª–∏–º–æ:

$ \binom{5}{3} = 10$$ $
$ 0.6^3 = 0.216 $
$ 0.4^2 = 0.16 $$ $

–¢–æ–º—É,

$ PX = 3 = 10 \cdot 0.216 \cdot 0.16 = 0.3456 $

–û—Ç–∂–µ, –π–º–æ–≤—ñ—Ä–Ω—ñ—Å—Ç—å —Ç–æ–≥–æ, —â–æ —Å–∞–º–µ 3 –∑ 5 –≤–∏–±—Ä–∞–Ω–∏—Ö —Ä–µ–π—Å—ñ–≤ –≤–∏–∫–æ–Ω—É—é—Ç—å—Å—è –Ω–∞ –≤–ª–∞—Å–Ω–æ–º—É –ø–∞—Ä–∫—É, –¥–æ—Ä—ñ–≤–Ω—é—î **0.3456** –∞–±–æ **34.56%**.


## –ó–∞–≤–¥–∞–Ω–Ω—è 3

–ü—ñ–¥–∫–∏–¥–∞—é—Ç—å —Ç—Ä–∏ –≥—Ä–∞–ª—å–Ω—ñ –∫—É–±–∏–∫–∏. –©–æ –π–º–æ–≤—ñ—Ä–Ω—ñ—à–µ: –æ—Ç—Ä–∏–º–∞—Ç–∏ –≤ —Å—É–º—ñ –æ—á–æ–∫, —â–æ –≤–∏–ø–∞–ª–∏, 11 –∞–±–æ 12? –†–æ–∑–≤‚Äô—è–∂–∏ –ø—Ä–æ–≥—Ä–∞–º–Ω–æ, –∑–≥–µ–Ω–µ—Ä—É–≤–∞–≤—à–∏ –≤—Å—ñ –º–æ–∂–ª–∏–≤—ñ –∫–æ–º–±—ñ–Ω–∞—Ü—ñ—ó.

```python
import itertools

# –ì–µ–Ω–µ—Ä—É—î–º–æ –≤—Å—ñ –º–æ–∂–ª–∏–≤—ñ –∫–æ–º–±—ñ–Ω–∞—Ü—ñ—ó –ø—ñ–¥–∫–∏–¥–∞–Ω–Ω—è —Ç—Ä—å–æ—Ö –∫—É–±–∏–∫—ñ–≤ (–≤—ñ–¥ 1 –¥–æ 6)
dice = range(1, 7)
combinations = list(itertools.product(dice, repeat=3))

# –õ—ñ—á–∏–ª—å–Ω–∏–∫–∏ –¥–ª—è —Å—É–º 11 —ñ 12
count_11 = 0
count_12 = 0

# –ü–µ—Ä–µ–±–∏—Ä–∞—î–º–æ –≤—Å—ñ –∫–æ–º–±—ñ–Ω–∞—Ü—ñ—ó
for combo in combinations:
    total = sum(combo)
    if total == 11:
        count_11 += 1
    elif total == 12:
        count_12 += 1

# –û–±—á–∏—Å–ª—é—î–º–æ –π–º–æ–≤—ñ—Ä–Ω–æ—Å—Ç—ñ
total_combinations = len(combinations)
prob_11 = count_11 / total_combinations
prob_12 = count_12 / total_combinations

# –í–∏–≤–æ–¥–∏–º–æ —Ä–µ–∑—É–ª—å—Ç–∞—Ç
print(f"–ö—ñ–ª—å–∫—ñ—Å—Ç—å –∫–æ–º–±—ñ–Ω–∞—Ü—ñ–π –¥–ª—è —Å—É–º–∏ 11: {count_11}")
print(f"–ö—ñ–ª—å–∫—ñ—Å—Ç—å –∫–æ–º–±—ñ–Ω–∞—Ü—ñ–π –¥–ª—è —Å—É–º–∏ 12: {count_12}")
print(f"–ó–∞–≥–∞–ª—å–Ω–∞ –∫—ñ–ª—å–∫—ñ—Å—Ç—å –∫–æ–º–±—ñ–Ω–∞—Ü—ñ–π: {total_combinations}")
print(f"–ô–º–æ–≤—ñ—Ä–Ω—ñ—Å—Ç—å –æ—Ç—Ä–∏–º–∞—Ç–∏ —Å—É–º—É 11: {prob_11:.4f}")
print(f"–ô–º–æ–≤—ñ—Ä–Ω—ñ—Å—Ç—å –æ—Ç—Ä–∏–º–∞—Ç–∏ —Å—É–º—É 12: {prob_12:.4f}")
```

### –ê–Ω–∞–ª—ñ—Ç–∏—á–Ω–µ —Ä—ñ—à–µ–Ω–Ω—è

1. **–ü–ª–æ—â–∞ –≤–µ–ª–∏–∫–æ–≥–æ –∫–æ–ª–∞:**

   –†–∞–¥—ñ—É—Å –≤–µ–ª–∏–∫–æ–≥–æ –∫–æ–ª–∞ $R = 10$ —Å–º.
   
   –ü–ª–æ—â–∞ –≤–µ–ª–∏–∫–æ–≥–æ –∫–æ–ª–∞ $S_{\text{–≤–µ–ª–∏–∫–µ}}$ –æ–±—á–∏—Å–ª—é—î—Ç—å—Å—è –∑–∞ —Ñ–æ—Ä–º—É–ª–æ—é:
   $
   S_{\text{–≤–µ–ª–∏–∫–µ}} = \pi R^2 = \pi \cdot 10^2 = 100\pi \, \text{—Å–º}^2
   $

2. **–ü–ª–æ—â–∞ –º–∞–ª–æ–≥–æ –∫–æ–ª–∞:**

   –†–∞–¥—ñ—É—Å –º–∞–ª–æ–≥–æ –∫–æ–ª–∞ $r = 5$ —Å–º.
   
   –ü–ª–æ—â–∞ –º–∞–ª–æ–≥–æ –∫–æ–ª–∞ $S_{\text{–º–∞–ª–µ}}$ –æ–±—á–∏—Å–ª—é—î—Ç—å—Å—è –∑–∞ —Ñ–æ—Ä–º—É–ª–æ—é:
   $
   S_{\text{–º–∞–ª–µ}} = \pi r^2 = \pi \cdot 5^2 = 25\pi \, \text{—Å–º}^2
   $

3. **–ü–ª–æ—â–∞ –∫—ñ–ª—å—Ü—è:**

   –ü–ª–æ—â–∞ –∫—ñ–ª—å—Ü—è $S_{\text{–∫—ñ–ª—å—Ü–µ}}$ –¥–æ—Ä—ñ–≤–Ω—é—î —Ä—ñ–∑–Ω–∏—Ü—ñ –ø–ª–æ—â –≤–µ–ª–∏–∫–æ–≥–æ –∫–æ–ª–∞ —ñ –º–∞–ª–æ–≥–æ –∫–æ–ª–∞:
   $
   S_{\text{–∫—ñ–ª—å—Ü–µ}} = S_{\text{–≤–µ–ª–∏–∫–µ}} - S_{\text{–º–∞–ª–µ}} = 100\pi - 25\pi = 75\pi \, \text{—Å–º}^2
   $

4. **–ô–º–æ–≤—ñ—Ä–Ω—ñ—Å—Ç—å:**

   –ô–º–æ–≤—ñ—Ä–Ω—ñ—Å—Ç—å —Ç–æ–≥–æ, —â–æ —Ç–æ—á–∫–∞, –∫–∏–Ω—É—Ç–∞ –Ω–∞–≤–º–∞–Ω–Ω—è —É –≤–µ–ª–∏–∫–µ –∫–æ–ª–æ, –ø–æ—Ç—Ä–∞–ø–∏—Ç—å —É –∫—ñ–ª—å—Ü–µ, –¥–æ—Ä—ñ–≤–Ω—é—î –≤—ñ–¥–Ω–æ—à–µ–Ω–Ω—é –ø–ª–æ—â—ñ –∫—ñ–ª—å—Ü—è –¥–æ –ø–ª–æ—â—ñ –≤–µ–ª–∏–∫–æ–≥–æ –∫–æ–ª–∞:
   $
   P = \frac{S_{\text{–∫—ñ–ª—å—Ü–µ}}}{S_{\text{–≤–µ–ª–∏–∫–µ}}} = \frac{75\pi}{100\pi} = \frac{75}{100} = 0.75
   $

### –í—ñ–¥–ø–æ–≤—ñ–¥—å

–ô–º–æ–≤—ñ—Ä–Ω—ñ—Å—Ç—å —Ç–æ–≥–æ, —â–æ —Ç–æ—á–∫–∞, –∫–∏–Ω—É—Ç–∞ –Ω–∞–≤–º–∞–Ω–Ω—è —É –≤–µ–ª–∏–∫–µ –∫–æ–ª–æ, –ø–æ—Ç—Ä–∞–ø–∏—Ç—å —Ç–∞–∫–æ–∂ —É –∫—ñ–ª—å—Ü–µ, —É—Ç–≤–æ—Ä–µ–Ω–µ –¥–≤–æ–º–∞ –∫–æ–Ω—Ü–µ–Ω—Ç—Ä–∏—á–Ω–∏–º–∏ –∫–æ–ª–∞–º–∏, –¥–æ—Ä—ñ–≤–Ω—é—î $0.75$ –∞–±–æ $75\%$.


## –ó–∞–≤–¥–∞–Ω–Ω—è 6

–£—Ä–Ω–∞ –º—ñ—Å—Ç–∏—Ç—å –æ–¥–Ω—É –∫—É–ª—é, –ø—Ä–æ —è–∫—É –≤—ñ–¥–æ–º–æ, —â–æ –≤–æ–Ω–∞ –∞–±–æ –±—ñ–ª–∞, –∞–±–æ —á–æ—Ä–Ω–∞ –∑ –æ–¥–Ω–∞–∫–æ–≤–∏–º–∏ –π–º–æ–≤—ñ—Ä–Ω–æ—Å—Ç—è–º–∏. –í —É—Ä–Ω—É –∫–ª–∞–¥—É—Ç—å –±—ñ–ª—É –∫—É–ª—é —ñ –ø–æ—Ç—ñ–º –Ω–∞–≤–º–∞–Ω–Ω—è –≤–∏–π–º–∞—é—Ç—å –æ–¥–Ω—É –∫—É–ª—é. –í–æ–Ω–∞ –≤–∏—è–≤–∏–ª–∞—Å—è –±—ñ–ª–æ—é. –Ø–∫–∞ –π–º–æ–≤—ñ—Ä–Ω—ñ—Å—Ç—å —Ç–æ–≥–æ, —â–æ –∫—É–ª—è, —è–∫–∞ –∑–∞–ª–∏—à–∏–ª–∞—Å—è, —î –±—ñ–ª–æ—é? –†–æ–∑–≤‚Äô—è–∂–∏ –∞–Ω–∞–ª—ñ—Ç–∏—á–Ω–æ.



### –ê–Ω–∞–ª—ñ—Ç–∏—á–Ω–µ —Ä—ñ—à–µ–Ω–Ω—è

1. **–í–∏—Ö—ñ–¥–Ω—ñ –¥–∞–Ω—ñ —Ç–∞ –π–º–æ–≤—ñ—Ä–Ω–æ—Å—Ç—ñ:**
   - –£—Ä–Ω–∞ –º—ñ—Å—Ç–∏—Ç—å –æ–¥–Ω—É –∫—É–ª—é, —è–∫–∞ –º–æ–∂–µ –±—É—Ç–∏ –±—ñ–ª–æ—é ($B$) –∞–±–æ —á–æ—Ä–Ω–æ—é ($C$) –∑ –æ–¥–Ω–∞–∫–æ–≤–∏–º–∏ –π–º–æ–≤—ñ—Ä–Ω–æ—Å—Ç—è–º–∏:
     
     P(B) = P(C) = 0.5
     
   - –î–æ–¥–∞—î–º–æ –≤ —É—Ä–Ω—É –±—ñ–ª—É –∫—É–ª—é, –ø—ñ—Å–ª—è —á–æ–≥–æ –≤ —É—Ä–Ω—ñ –¥–≤—ñ –∫—É–ª—ñ:
     - –£ –ø–µ—Ä—à–æ–º—É –≤–∏–ø–∞–¥–∫—É: –¥–≤—ñ –±—ñ–ª—ñ –∫—É–ª—ñ (BB).
     - –£ –¥—Ä—É–≥–æ–º—É –≤–∏–ø–∞–¥–∫—É: –æ–¥–Ω–∞ –±—ñ–ª–∞ —ñ –æ–¥–Ω–∞ —á–æ—Ä–Ω–∞ –∫—É–ª—è (BC).

2. **–ô–º–æ–≤—ñ—Ä–Ω—ñ—Å—Ç—å –≤–∏—Ç—è–≥–Ω—É—Ç–∏ –±—ñ–ª—É –∫—É–ª—é:**
   - –Ø–∫—â–æ –≤ —É—Ä–Ω—ñ –¥–≤—ñ –±—ñ–ª—ñ –∫—É–ª—ñ (BB):
     $
     P(\text{–±—ñ–ª—É} \mid BB) = 1
     $
   - –Ø–∫—â–æ –≤ —É—Ä–Ω—ñ –æ–¥–Ω–∞ –±—ñ–ª–∞ —ñ –æ–¥–Ω–∞ —á–æ—Ä–Ω–∞ –∫—É–ª—è (BC):
     $
     P(\text{–±—ñ–ª—É} \mid BC) = 0.5
     $

3. **–ó–∞–≥–∞–ª—å–Ω–∞ –π–º–æ–≤—ñ—Ä–Ω—ñ—Å—Ç—å –≤–∏—Ç—è–≥–Ω—É—Ç–∏ –±—ñ–ª—É –∫—É–ª—é:**
   $
   P(\text{–±—ñ–ª—É}) = P(\text{–±—ñ–ª—É} \mid BB) \cdot P(B) + P(\text{–±—ñ–ª—É} \mid BC) \cdot P(C)
   $
   $
   P(\text{–±—ñ–ª—É}) = 1 \cdot 0.5 + 0.5 \cdot 0.5 = 0.5 + 0.25 = 0.75
   $

4. **–ô–º–æ–≤—ñ—Ä–Ω—ñ—Å—Ç—å —Ç–æ–≥–æ, —â–æ –∑–∞–ª–∏—à–∏–ª–∞—Å—è –±—ñ–ª–∞ –∫—É–ª—è:**
   - –í–∏–∑–Ω–∞—á–∏–º–æ –π–º–æ–≤—ñ—Ä–Ω—ñ—Å—Ç—å —Ç–æ–≥–æ, —â–æ –ø–æ—á–∞—Ç–∫–æ–≤–∞ –∫—É–ª—è –±—É–ª–∞ –±—ñ–ª–æ—é, –∑–∞ —É–º–æ–≤–∏, —â–æ –≤–∏—Ç—è–≥–Ω—É—Ç–∞ –±—ñ–ª–∞ –∫—É–ª—è:
     $
     P(B \mid \text{–±—ñ–ª—É}) = \frac{P(\text{–±—ñ–ª—É} \mid B) \cdot P(B)}{P(\text{–±—ñ–ª—É})}
     $
   - –ó –ø—ñ–¥—Å—Ç–∞–≤–ª–µ–Ω–Ω—è–º:
     $
     P(B \mid \text{–±—ñ–ª—É}) = \frac{1 \cdot 0.5}{0.75} = \frac{0.5}{0.75} = \frac{2}{3}
     $

5. **–ô–º–æ–≤—ñ—Ä–Ω—ñ—Å—Ç—å —Ç–æ–≥–æ, —â–æ –∑–∞–ª–∏—à–∏–ª–∞—Å—è –±—ñ–ª–∞ –∫—É–ª—è, —è–∫—â–æ –≤–∏—Ç—è–≥–Ω—É–ª–∏ –±—ñ–ª—É:**
   - –Ø–∫—â–æ –ø–æ—á–∞—Ç–∫–æ–≤–∞ –∫—É–ª—è –±—É–ª–∞ –±—ñ–ª–æ—é, –¥—Ä—É–≥–∞ –∫—É–ª—è —Ç–∞–∫–æ–∂ –±—ñ–ª–∞ –∑ –π–º–æ–≤—ñ—Ä–Ω—ñ—Å—Ç—é 1:
     $
     P(\text{–±—ñ–ª–∞ –∑–∞–ª–∏—à–∏–ª–∞—Å—è} \mid B) = 1
     $
   - –Ø–∫—â–æ –ø–æ—á–∞—Ç–∫–æ–≤–∞ –∫—É–ª—è –±—É–ª–∞ —á–æ—Ä–Ω–æ—é, –∑–∞–ª–∏—à–∏–ª–∞—Å—è –æ–¥–Ω–∞ —á–æ—Ä–Ω–∞ –∫—É–ª—è:
     $
     P(\text{–±—ñ–ª–∞ –∑–∞–ª–∏—à–∏–ª–∞—Å—è} \mid C) = 0
     $

   –í—Ä–∞—Ö–æ–≤—É—é—á–∏ –π–º–æ–≤—ñ—Ä–Ω—ñ—Å—Ç—å, —â–æ –ø–æ—á–∞—Ç–∫–æ–≤–∞ –∫—É–ª—è –±—É–ª–∞ –±—ñ–ª–æ—é:
   $
   P(\text{–±—ñ–ª–∞ –∑–∞–ª–∏—à–∏–ª–∞—Å—è} \mid \text{–±—ñ–ª—É}) = \frac{2}{3}
   $

### –í—ñ–¥–ø–æ–≤—ñ–¥—å

–ô–º–æ–≤—ñ—Ä–Ω—ñ—Å—Ç—å —Ç–æ–≥–æ, —â–æ –∫—É–ª—è, —è–∫–∞ –∑–∞–ª–∏—à–∏–ª–∞—Å—è, —î –±—ñ–ª–æ—é, –¥–æ—Ä—ñ–≤–Ω—é—î $\frac{2}{3}$ –∞–±–æ –ø—Ä–∏–±–ª–∏–∑–Ω–æ $66.67\%$.


## –ó–∞–≤–¥–∞–Ω–Ω—è 7. –í–∏–ø–∞–¥–∫–æ–≤–µ –±–ª—É–∫–∞–Ω–Ω—è

–¶—ñ–Ω–∞ –∞–∫—Ü—ñ–π –∫–æ–º–ø–∞–Ω—ñ—ó –≤ –∫–æ–∂–µ–Ω –º–æ–º–µ–Ω—Ç —á–∞—Å—É –º–æ–∂–µ –∑ —Ä—ñ–≤–Ω–æ—é –π–º–æ–≤—ñ—Ä–Ω—ñ—Å—Ç—é –∑–±—ñ–ª—å—à–∏—Ç–∏—Å—å –Ω–∞ 2 –∞–±–æ –∑–º–µ–Ω—à–∏—Ç–∏—Å—å –Ω–∞ 1. –£ –ø–æ—á–∞—Ç–∫–æ–≤–∏–π –º–æ–º–µ–Ω—Ç —á–∞—Å—É
ùë°

t=0 —Ü—ñ–Ω–∞ —Ä—ñ–≤–Ω–∞ 0. –í–∏–∑–Ω–∞—á —Å–µ—Ä–µ–¥–Ω—é —Ü—ñ–Ω—É –∞–∫—Ü—ñ—ó —á–µ—Ä–µ–∑ 3 –æ–¥–∏–Ω–∏—Ü—ñ —á–∞—Å—É (

–∞) –†–æ–∑–≤‚Äô—è–∂–∏ –∞–Ω–∞–ª—ñ—Ç–∏—á–Ω–æ.

–±) –ù–∞–ø–∏—à–∏ —Å–∏–º—É–ª—è—Ü—ñ—é –¥–∞–Ω–æ–≥–æ –ø—Ä–æ—Ü–µ—Å—É. –ü–æ—Ä–∞—Ö—É–π —Å–µ—Ä–µ–¥–Ω—î –∑–Ω–∞—á–µ–Ω–Ω—è —Ç–∞ –Ω–∞–º–∞–ª—é–π –≥—ñ—Å—Ç–æ–≥—Ä–∞–º—É —Ü—ñ–Ω–∏ –¥–ª—è 10, 100, 1000 —Ç–∞ 10000 —Å–∏–º—É–ª—è—Ü—ñ–π.



–ü–æ–∑–Ω–∞—á–∏–º–æ:
- $S_t$  ‚Äî —Ü—ñ–Ω–∞ –∞–∫—Ü—ñ–π —É –º–æ–º–µ–Ω—Ç —á–∞—Å—É  $t$ .
- $ S_0 = 0 $.

–ù–∞ –∫–æ–∂–Ω–æ–º—É –∫—Ä–æ—Ü—ñ —á–∞—Å—É —Ü—ñ–Ω–∞ –∑–º—ñ–Ω—é—î—Ç—å—Å—è —Ç–∞–∫–∏–º —á–∏–Ω–æ–º:
$
S_{t+1} = S_t + X_t
$
–¥–µ $ X_t $ ‚Äî –≤–∏–ø–∞–¥–∫–æ–≤–∞ –∑–º—ñ–Ω–Ω–∞, —è–∫–∞ –ø—Ä–∏–π–º–∞—î –∑–Ω–∞—á–µ–Ω–Ω—è +2 –∑ –π–º–æ–≤—ñ—Ä–Ω—ñ—Å—Ç—é 0.5 —ñ -1 –∑ –π–º–æ–≤—ñ—Ä–Ω—ñ—Å—Ç—é 0.5.

–°–µ—Ä–µ–¥–Ω—î –∑–Ω–∞—á–µ–Ω–Ω—è $ X_t $:
$
\mathbb{E}[X_t] = 0.5 \times 2 + 0.5 \times (-1) = 1 - 0.5 = 0.5
$

–ú–∞—Ç–µ–º–∞—Ç–∏—á–Ω–µ —Å–ø–æ–¥—ñ–≤–∞–Ω–Ω—è —î –ª—ñ–Ω—ñ–π–Ω–∏–º, —Ç–æ–º—É:
$
\mathbb{E}[S_{t+1}] = \mathbb{E}[S_t + X_t] = \mathbb{E}[S_t] + \mathbb{E}[X_t]
$

–û—Å–∫—ñ–ª—å–∫–∏ \( S_0 = 0 \), –æ—Ç—Ä–∏–º–∞—î–º–æ:
$
\mathbb{E}[S_1] = \mathbb{E}[S_0] + \mathbb{–ï}[X_0] = 0 + 0.5 = 0.5
$

–î–∞–ª—ñ:
$
\mathbb{–ï}[S_2] = \mathbb{–ï}[S_1] + \mathbb{–ï}[X_1] = 0.5 + 0.5 = 1
$
$
\mathbb{–ï}[S_3] = \mathbb{–ï}[S_2] + \mathbb{–ï}[X_2] = 1 + 0.5 = 1.5
$

–¢–∞–∫–∏–º —á–∏–Ω–æ–º, —Å–µ—Ä–µ–¥–Ω—è —Ü—ñ–Ω–∞ –∞–∫—Ü—ñ–π —á–µ—Ä–µ–∑ 3 –æ–¥–∏–Ω–∏—Ü—ñ —á–∞—Å—É:
$
\mathbb{–ï}[S_3] = 1.5
$

–û—Ç–∂–µ, –∞–Ω–∞–ª—ñ—Ç–∏—á–Ω–∏–π —Ä–æ–∑–≤'—è–∑–æ–∫ –ø–æ–∫–∞–∑—É—î, —â–æ —Å–µ—Ä–µ–¥–Ω—è —Ü—ñ–Ω–∞ –∞–∫—Ü—ñ–π —á–µ—Ä–µ–∑ 3 –æ–¥–∏–Ω–∏—Ü—ñ —á–∞—Å—É –¥–æ—Ä—ñ–≤–Ω—é—î 1.5.




```python
import numpy as np
import matplotlib.pyplot as plt

def simulate_price(steps):
    price = 0
    for _ in range(steps):
        price += np.random.choice([2, -1])
    return price

def run_simulations(num_simulations, steps):
    prices = [simulate_price(steps) for _ in range(num_simulations)]
    return prices

def plot_histogram(prices, num_simulations):
    plt.hist(prices, bins=range(min(prices), max(prices) + 1, 1), alpha=0.75)
    plt.title(f'–ì—ñ—Å—Ç–æ–≥—Ä–∞–º–∞ —Ü—ñ–Ω –ø—ñ—Å–ª—è {steps} –æ–¥–∏–Ω–∏—Ü—å —á–∞—Å—É ({num_simulations} —Å–∏–º—É–ª—è—Ü—ñ–π)')
    plt.xlabel('–¶—ñ–Ω–∞ –∞–∫—Ü—ñ–π')
    plt.ylabel('–ß–∞—Å—Ç–æ—Ç–∞')
    plt.grid(True)
    plt.show()

steps = 3

# 10 —Å–∏–º—É–ª—è—Ü—ñ–π
num_simulations = 10
prices = run_simulations(num_simulations, steps)
print(f'–°–µ—Ä–µ–¥–Ω—è —Ü—ñ–Ω–∞ –¥–ª—è {num_simulations} —Å–∏–º—É–ª—è—Ü—ñ–π: {np.mean(prices)}')
plot_histogram(prices, num_simulations)

# 100 —Å–∏–º—É–ª—è—Ü—ñ–π
num_simulations = 100
prices = run_simulations(num_simulations, steps)
print(f'–°–µ—Ä–µ–¥–Ω—è —Ü—ñ–Ω–∞ –¥–ª—è {num_simulations} —Å–∏–º—É–ª—è—Ü—ñ–π: {np.mean(prices)}')
plot_histogram(prices, num_simulations)

# 1000 —Å–∏–º—É–ª—è—Ü—ñ–π
num_simulations = 1000
prices = run_simulations(num_simulations, steps)
print(f'–°–µ—Ä–µ–¥–Ω—è —Ü—ñ–Ω–∞ –¥–ª—è {num_simulations} —Å–∏–º—É–ª—è—Ü—ñ–π: {np.mean(prices)}')
plot_histogram(prices, num_simulations)

# 10000 —Å–∏–º—É–ª—è—Ü—ñ–π
num_simulations = 10000
prices = run_simulations(num_simulations, steps)
print(f'–°–µ—Ä–µ–¥–Ω—è —Ü—ñ–Ω–∞ –¥–ª—è {num_simulations} —Å–∏–º—É–ª—è—Ü—ñ–π: {np.mean(prices)}')
plot_histogram(prices, num_simulations)
```
