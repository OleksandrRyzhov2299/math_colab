### Завдання 1: Ймовірність вибору нестандартних деталей

У партії з 50 деталей 5 нестандартних. Знайдемо ймовірність того, що серед обраних навмання для перевірки шести деталей дві виявляться нестандартними.

#### Розв'язання:

Ймовірність події обчислюється як відношення кількості сприятливих випадків до загальної кількості можливих випадків:

1. **Кількість сприятливих випадків**:
   Для того, щоб 2 із 6 обраних деталей були нестандартними, спершу потрібно вибрати 2 нестандартні деталі із наявних 5. Це можна зробити за допомогою комбінації $ \binom{5}{2} $.

2. **Кількість всіх можливих випадків**:
   Вибір 6 деталей із 50 можливий за допомогою комбінації $ \binom{50}{6} $.

Отже, ймовірність того, що серед обраних наугад шести деталей дві будуть нестандартними, обчислюється як:

$ P = \frac{\binom{5}{2}}{\binom{50}{6}} $

Де $ \binom{5}{2} $ - це число способів вибрати 2 нестандартні деталі з 5, а $ \binom{50}{6} $ - число способів вибрати будь-які 6 деталей з 50.

Обчисливши $ \binom{5}{2} = 10 $ та $ \binom{50}{6} $, можна знайти ймовірність $ P $.

Цей підхід дозволяє аналітично знайти точну ймовірність без потреби у конкретних числових значеннях.

Для задачі з вибіркою 5 рейсів із 15-ти, де 60% рейсів виконуються на власному парку, шукаємо ймовірність, що саме 3 з цих 5 рейсів виконуються на власному парку.

Маємо:
- Кількість рейсів: $ n = 15$
- Ймовірність рейсу на власному парку:  $p = 0.6$
- Кількість вибраних рейсів:  $k = 5 $
- Шукаємо ймовірність  $$PX = 3$$ , де \ X \ — кількість рейсів на власному парку серед 5-ти.

Використовуємо біноміальний розподіл:

 $ PX = 3 = \binom{5}{3} \cdot 0.6^3 \cdot 0.4^2 \ $

Обчислимо:

$ \binom{5}{3} = 10$$ $
$ 0.6^3 = 0.216 $
$ 0.4^2 = 0.16 $$ $

Тому,

$ PX = 3 = 10 \cdot 0.216 \cdot 0.16 = 0.3456 $

Отже, ймовірність того, що саме 3 з 5 вибраних рейсів виконуються на власному парку, дорівнює **0.3456** або **34.56%**.


## Завдання 3

Підкидають три гральні кубики. Що ймовірніше: отримати в сумі очок, що випали, 11 або 12? Розв’яжи програмно, згенерувавши всі можливі комбінації.

```python
import itertools

# Генеруємо всі можливі комбінації підкидання трьох кубиків (від 1 до 6)
dice = range(1, 7)
combinations = list(itertools.product(dice, repeat=3))

# Лічильники для сум 11 і 12
count_11 = 0
count_12 = 0

# Перебираємо всі комбінації
for combo in combinations:
    total = sum(combo)
    if total == 11:
        count_11 += 1
    elif total == 12:
        count_12 += 1

# Обчислюємо ймовірності
total_combinations = len(combinations)
prob_11 = count_11 / total_combinations
prob_12 = count_12 / total_combinations

# Виводимо результат
print(f"Кількість комбінацій для суми 11: {count_11}")
print(f"Кількість комбінацій для суми 12: {count_12}")
print(f"Загальна кількість комбінацій: {total_combinations}")
print(f"Ймовірність отримати суму 11: {prob_11:.4f}")
print(f"Ймовірність отримати суму 12: {prob_12:.4f}")
```

1. Елементи 1 і 2 з'єднані паралельно. \\
2. Елементи 3 і 4 з'єднані послідовно. \\
3. Елемент 5 з'єднаний послідовно з блоком (3 і 4). \\
4. Елемент 6 з'єднаний послідовно з блоком (1 і 2) і з блоком (3, 4, 5). \\

Для виходу з ладу всієї схеми (подія $ A $), необхідно, щоб хоча б один елемент з кожного послідовного з'єднання вийшов з ладу.

1. Для паралельного з'єднання (елементи 1 і 2):
   Блок вийде з ладу, якщо обидва елементи вийдуть з ладу: $ A_1 \land A_2 $.

2. Для послідовного з'єднання (елементи 3 і 4):
   Блок вийде з ладу, якщо хоча б один з елементів вийде з ладу: $ A_3 \lor A_4 $.

3. Для послідовного з'єднання (блок з елементів 3 і 4 з елементом 5):
   Блок вийде з ладу, якщо хоча б один елемент з цієї ланки вийде з ладу: $ (A_3 \lor A_4) \lor A_5 $.

Подія $ A $:
$$
A = (A_1 \land A_2) \land ((A_3 \lor A_4) \lor A_5) \land A_6
$$

Формула для доповнення події $ A $ ($ \overline{A} $):
$$
\overline{A} = (\overline{A_1} \lor \overline{A_2}) \lor ((\overline{A_3} \land \overline{A_4}) \lor \overline{A_5}) \lor \overline{A_6}
$$

Підсумок:

- Подія $ A $:
$$
A = (A_1 \land A_2) \land ((A_3 \lor A_4) \lor A_5) \land A_6
$$

- Доповнення $ \overline{A} $:
$$
\overline{A} = (\overline{A_1} \lor \overline{A_2}) \lor ((\overline{A_3} \land \overline{A_4}) \lor \overline{A_5}) \lor \overline{A_6}
$$

Ці формули виражають ймовірність виходу з ладу всієї схеми та її доповнення через події виходу з ладу окремих елементів.

На площині накреслені дві концентричні окружності (мають спільний центр), радіуси яких 5 і 10 см відповідно. Знайди ймовірність того, що точка, кинута навмання у велике коло, потрапить також у кільце, утворене побудованими окружностями. Розв’яжи аналітично.

Для вирішення завдання знайдемо ймовірність того, що точка, кинута навмання у велике коло радіусом 10 см, потрапить також у кільце, утворене двома концентричними колами з радіусами 5 см і 10 см.

### Аналітичне рішення

1. **Площа великого кола:**

   Радіус великого кола $R = 10$ см.
   
   Площа великого кола $S_{\text{велике}}$ обчислюється за формулою:
   $
   S_{\text{велике}} = \pi R^2 = \pi \cdot 10^2 = 100\pi \, \text{см}^2
   $

2. **Площа малого кола:**

   Радіус малого кола $r = 5$ см.
   
   Площа малого кола $S_{\text{мале}}$ обчислюється за формулою:
   $
   S_{\text{мале}} = \pi r^2 = \pi \cdot 5^2 = 25\pi \, \text{см}^2
   $

3. **Площа кільця:**

   Площа кільця $S_{\text{кільце}}$ дорівнює різниці площ великого кола і малого кола:
   $
   S_{\text{кільце}} = S_{\text{велике}} - S_{\text{мале}} = 100\pi - 25\pi = 75\pi \, \text{см}^2
   $

4. **Ймовірність:**

   Ймовірність того, що точка, кинута навмання у велике коло, потрапить у кільце, дорівнює відношенню площі кільця до площі великого кола:
   $
   P = \frac{S_{\text{кільце}}}{S_{\text{велике}}} = \frac{75\pi}{100\pi} = \frac{75}{100} = 0.75
   $

### Відповідь

Ймовірність того, що точка, кинута навмання у велике коло, потрапить також у кільце, утворене двома концентричними колами, дорівнює $0.75$ або $75\%$.


## Завдання 6

Урна містить одну кулю, про яку відомо, що вона або біла, або чорна з однаковими ймовірностями. В урну кладуть білу кулю і потім навмання виймають одну кулю. Вона виявилася білою. Яка ймовірність того, що куля, яка залишилася, є білою? Розв’яжи аналітично.



### Аналітичне рішення

1. **Вихідні дані та ймовірності:**
   - Урна містить одну кулю, яка може бути білою ($B$) або чорною ($C$) з однаковими ймовірностями:
     
     P(B) = P(C) = 0.5
     
   - Додаємо в урну білу кулю, після чого в урні дві кулі:
     - У першому випадку: дві білі кулі (BB).
     - У другому випадку: одна біла і одна чорна куля (BC).

2. **Ймовірність витягнути білу кулю:**
   - Якщо в урні дві білі кулі (BB):
     $
     P(\text{білу} \mid BB) = 1
     $
   - Якщо в урні одна біла і одна чорна куля (BC):
     $
     P(\text{білу} \mid BC) = 0.5
     $

3. **Загальна ймовірність витягнути білу кулю:**
   $
   P(\text{білу}) = P(\text{білу} \mid BB) \cdot P(B) + P(\text{білу} \mid BC) \cdot P(C)
   $
   $
   P(\text{білу}) = 1 \cdot 0.5 + 0.5 \cdot 0.5 = 0.5 + 0.25 = 0.75
   $

4. **Ймовірність того, що залишилася біла куля:**
   - Визначимо ймовірність того, що початкова куля була білою, за умови, що витягнута біла куля:
     $
     P(B \mid \text{білу}) = \frac{P(\text{білу} \mid B) \cdot P(B)}{P(\text{білу})}
     $
   - З підставленням:
     $
     P(B \mid \text{білу}) = \frac{1 \cdot 0.5}{0.75} = \frac{0.5}{0.75} = \frac{2}{3}
     $

5. **Ймовірність того, що залишилася біла куля, якщо витягнули білу:**
   - Якщо початкова куля була білою, друга куля також біла з ймовірністю 1:
     $
     P(\text{біла залишилася} \mid B) = 1
     $
   - Якщо початкова куля була чорною, залишилася одна чорна куля:
     $
     P(\text{біла залишилася} \mid C) = 0
     $

   Враховуючи ймовірність, що початкова куля була білою:
   $
   P(\text{біла залишилася} \mid \text{білу}) = \frac{2}{3}
   $

### Відповідь

Ймовірність того, що куля, яка залишилася, є білою, дорівнює $\frac{2}{3}$ або приблизно $66.67\%$.

Ціна акцій компанії в кожен момент часу може з рівною ймовірністю збільшитись на 2 або зменшитись на 1. У початковий момент часу

а) Розв’яжи аналітично.

б) Напиши симуляцію даного процесу. Порахуй середнє значення та намалюй гістограму ціни для 10, 100, 1000 та 10000 симуляцій.



Позначимо:
- $S_t$  — ціна акцій у момент часу  $t$ .
- $ S_0 = 0 $.

На кожному кроці часу ціна змінюється таким чином:
$
S_{t+1} = S_t + X_t
$
де $ X_t $ — випадкова змінна, яка приймає значення +2 з ймовірністю 0.5 і -1 з ймовірністю 0.5.

Середнє значення $ X_t $:
$
\mathbb{E}[X_t] = 0.5 \times 2 + 0.5 \times (-1) = 1 - 0.5 = 0.5
$

Математичне сподівання є лінійним, тому:
$
\mathbb{E}[S_{t+1}] = \mathbb{E}[S_t + X_t] = \mathbb{E}[S_t] + \mathbb{E}[X_t]
$

Оскільки \( S_0 = 0 \), отримаємо:
$
\mathbb{E}[S_1] = \mathbb{E}[S_0] + \mathbb{Е}[X_0] = 0 + 0.5 = 0.5
$

Далі:
$
\mathbb{Е}[S_2] = \mathbb{Е}[S_1] + \mathbb{Е}[X_1] = 0.5 + 0.5 = 1
$
$
\mathbb{Е}[S_3] = \mathbb{Е}[S_2] + \mathbb{Е}[X_2] = 1 + 0.5 = 1.5
$

Таким чином, середня ціна акцій через 3 одиниці часу:
$
\mathbb{Е}[S_3] = 1.5
$

Отже, аналітичний розв'язок показує, що середня ціна акцій через 3 одиниці часу дорівнює 1.5.




```python
import numpy as np
import matplotlib.pyplot as plt

def simulate_price(steps):
    price = 0
    for _ in range(steps):
        price += np.random.choice([2, -1])
    return price

def run_simulations(num_simulations, steps):
    prices = [simulate_price(steps) for _ in range(num_simulations)]
    return prices

def plot_histogram(prices, num_simulations):
    plt.hist(prices, bins=range(min(prices), max(prices) + 1, 1), alpha=0.75)
    plt.title(f'Гістограма цін після {steps} одиниць часу ({num_simulations} симуляцій)')
    plt.xlabel('Ціна акцій')
    plt.ylabel('Частота')
    plt.grid(True)
    plt.show()

steps = 3

# 10 симуляцій
num_simulations = 10
prices = run_simulations(num_simulations, steps)
print(f'Середня ціна для {num_simulations} симуляцій: {np.mean(prices)}')
plot_histogram(prices, num_simulations)

# 100 симуляцій
num_simulations = 100
prices = run_simulations(num_simulations, steps)
print(f'Середня ціна для {num_simulations} симуляцій: {np.mean(prices)}')
plot_histogram(prices, num_simulations)

# 1000 симуляцій
num_simulations = 1000
prices = run_simulations(num_simulations, steps)
print(f'Середня ціна для {num_simulations} симуляцій: {np.mean(prices)}')
plot_histogram(prices, num_simulations)

# 10000 симуляцій
num_simulations = 10000
prices = run_simulations(num_simulations, steps)
print(f'Середня ціна для {num_simulations} симуляцій: {np.mean(prices)}')
plot_histogram(prices, num_simulations)
```
