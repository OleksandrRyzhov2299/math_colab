## Завдання 1
Візуалізуй функцію на відрізку від 0 до 24. За допомогою SymPy порахуй невизначений інтеграл функції та інтеграл від a до b.

```python
import numpy as np
import matplotlib.pyplot as plt
from sympy import symbols, exp, pi, sqrt, integrate

# Определяем переменную
x = symbols('x')

# Определяем функцию
term1 = (4 / (1.2 * sqrt(2 * pi))) * exp(-0.5 * ((x - 11) / 1.2) ** 2)
term2 = (7 / (2.4 * sqrt(2 * pi))) * exp(-0.5 * ((x - 15) / 2.4) ** 2)
function = 2 * (term1 + term2)

# 1. Визуализация функции на интервале от 0 до 24
x_vals = np.linspace(0, 24, 1000)
y_vals = [function.evalf(subs={x: val}) for val in x_vals]

plt.plot(x_vals, y_vals, label='Taxi Efficiency Function')
plt.xlabel('Time (hours)')
plt.ylabel('Efficiency')
plt.title('Taxi Efficiency Over Time')
plt.legend()
plt.grid(True)
plt.show()

# 2. Вычисление неопределенного интеграла функции с помощью SymPy
indefinite_integral = integrate(function, x)
print(f"Неопределенный интеграл функции: {indefinite_integral}")

# 3. Вычисление определенного интеграла от a до b с помощью SymPy
a, b = 9, 18
definite_integral = integrate(function, (x, a, b))
print(f"Определенный интеграл функции от {a} до {b}: {definite_integral}")
```

## Завдання 2
Напиши функцію чисельного інтегрування методом прямокутників та порахуй інтеграл від a до b

```python
import numpy as np
from sympy import symbols, exp, pi, sqrt

# Определяем переменную
x = symbols('x')

# Определяем функцию
term1 = (4 / (1.2 * sqrt(2 * pi))) * exp(-0.5 * ((x - 11) / 1.2) ** 2)
term2 = (7 / (2.4 * sqrt(2 * pi))) * exp(-0.5 * ((x - 15) / 2.4) ** 2)
function = 2 * (term1 + term2)

# Преобразуем sympy функцию в lambda для использования в численных методах
f_lambda = lambda x_val: function.evalf(subs={x: x_val})

# Функция численного интегрирования методом прямоугольников
def rectangle_method(f, a, b, n):
    h = (b - a) / n
    result = sum(f(a + i * h) for i in range(n)) * h
    return result

# Определяем границы интегрирования
a, b = 9, 18
n = 1000  # Количество разбиений

# Вычисляем интеграл методом прямоугольников
rect_integral = rectangle_method(f_lambda, a, b, n)
print(f"Интеграл методом прямоугольников от {a} до {b}: {rect_integral}")
```

## Завдання 3
Напиши функцію чисельного інтегрування методом трапецій та порахуй інтеграл від a до b.

```python
import numpy as np
from sympy import symbols, exp, pi, sqrt

# Определяем переменную
x = symbols('x')

# Определяем функцию
term1 = (4 / (1.2 * sqrt(2 * pi))) * exp(-0.5 * ((x - 11) / 1.2) ** 2)
term2 = (7 / (2.4 * sqrt(2 * pi))) * exp(-0.5 * ((x - 15) / 2.4) ** 2)
function = 2 * (term1 + term2)

# Преобразуем sympy функцию в lambda для использования в численных методах
f_lambda = lambda x_val: function.evalf(subs={x: x_val})

# Функция численного интегрирования методом трапеций
def trapezoid_method(f, a, b, n):
    h = (b - a) / n
    result = 0.5 * (f(a) + f(b)) + sum(f(a + i * h) for i in range(1, n))
    return result * h

# Определяем границы интегрирования
a, b = 9, 18
n = 1000  # Количество разбиений

# Вычисляем интеграл методом трапеций
trap_integral = trapezoid_method(f_lambda, a, b, n)
print(f"Интеграл методом трапеций от {a} до {b}: {trap_integral}")
```

## Завдання 4
Напиши функцію чисельного інтегрування методом Сімпсона та порахуй інтеграл від a до b.

```python
import numpy as np
from sympy import symbols, exp, pi, sqrt

# Определяем переменную
x = symbols('x')

# Определяем функцию
term1 = (4 / (1.2 * sqrt(2 * pi))) * exp(-0.5 * ((x - 11) / 1.2) ** 2)
term2 = (7 / (2.4 * sqrt(2 * pi))) * exp(-0.5 * ((x - 15) / 2.4) ** 2)
function = 2 * (term1 + term2)

# Преобразуем sympy функцию в lambda для использования в численных методах
f_lambda = lambda x_val: function.evalf(subs={x: x_val})

# Функция численного интегрирования методом Симпсона
def simpson_method(f, a, b, n):
    if n % 2 == 1:
        n += 1  # Simpson's rule requires an even number of intervals
    h = (b - a) / n
    result = f(a) + f(b) + 4 * sum(f(a + i * h) for i in range(1, n, 2)) + 2 * sum(f(a + i * h) for i in range(2, n-1, 2))
    return result * h / 3

# Определяем границы интегрирования
a, b = 9, 18
n = 1000  # Количество разбиений

# Вычисляем интеграл методом Симпсона
simp_integral = simpson_method(f_lambda, a, b, n)
print(f"Интеграл методом Симпсона от {a} до {b}: {simp_integral}")
```
